```python
import requests
from bs4 import BeautifulSoup
```

```python
def retrieve_name(id_):
    try:
        name = soup.find('div', {'id':id_}).find('div', {'class': 'name'}).find('span', {'itemprop': 'name'})
        print(name.get_text())
        return name
    except:
        None
```

```python
def retrieve_publisher(id_):
    try:
        publisher = soup.find('div', {'id':id_}).find('div', {'class': 'publisher'}).find('span', {'itemprop': 'name'})
        print(publisher.get_text())
        return publisher
    except:
        None
```

```python
def retrieve_author(id_):
    try:
        author = soup.find('div', {'id':id_}).find('div', {'class': 'author'}).find('span', {'itemprop': 'name'})
        print(author.get_text())
        return author
    except:
        None
```

```python
def retrieve_price(id_):
    try:
        price = soup.find('div', {'id':id_}).find('div', {'class': 'price-new'}).find('span', {'class': 'value'})
        print(price.get_text())
        return price
    except:
        price = soup.find('div', {'id':id_}).find('span', {'class': 'price-old'}).find('span', {'class': 'value'})
        print(price.get_text())
```

```python
def retrieve_ids():
    lst = []
    all_id = soup.find_all('div', {'itemtype':'http://schema.org/Book'})
    for single_id in all_id:
        lst.append(single_id.attrs['id'])
    return lst
```

```python
def retrieve_all():
    for id_ in retrieve_ids():
        retrieve_name(id_)
        retrieve_publisher(id_)
        retrieve_author(id_)
        retrieve_price(id_)
```
